```python
from sympy import symbols, integrate
import sympy
from IPython.display import display,Math
import numpy as np
x = symbols('x')
# 计算 ∫x² dx
result = integrate(x**2, x)
print(result)  # 输出: x**3/3
```

```python
# f(x) = x^2
# F(x) = (1/3)x^3
# find: area from [1,2]
```

```python
true_area = (1/3) *2**3 - (1/3)* 1**3
```

```python
estimated_area = 0
x_values = np.linspace(1,2,1000)
for i in range(len(x_values) -1):
    x1 = x_values[i]
    x2 = x_values[i+1]
    width = x2 - x1
    x_mid = (x1 + x2) / 2
    height = x_mid **2
    estimated_area += width * height

print('true area:',true_area)
print('estimated area:', estimated_area)
```

```python
# f(x) = sin(x)
# F(x) = -cos(x)
# find: area from [0,pi]

true_area = -np.cos(np.pi) - (-np.cos(0))
estimated_area = 0
x_values = np.linspace(0,np.pi, 1000)
for i in range(len(x_values) -1):
    x1 = x_values[i]
    x2 = x_values[i+1]
    width = x2 - x1
    x_mid = (x1 + x2) / 2
    height = np.sin(x_mid)
    estimated_area += width * height
print('true area:',true_area)
print('estimated area:', estimated_area)
```
