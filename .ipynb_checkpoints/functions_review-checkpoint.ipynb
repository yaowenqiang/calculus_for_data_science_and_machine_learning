```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,100)
y = 0.5 * x + 1
plt.axhline(y=0,color='k')
plt.axvline(x=0,color='k')
plt.grid()
plt.plot(x,y)
```

```python
x = np.linspace(-2,2,100)
y = -0.5*9.8*x**2+2*x + 1
plt.axhline(y=0,color='k')
plt.axvline(x=0,color='k')
plt.grid()
plt.plot(x,y)
```

```python
x = np.linspace(0,10,100)
y = -0.5*x**(1./3) + 0.2*x**(1./2) + 1
plt.axhline(y=0,color='k')
plt.axvline(x=0,color='k')
plt.grid()
plt.plot(x,y)
```

```python
x = np.linspace(-5,5,100)
y = np.exp(x) #e**x
plt.axhline(y=0,color='k')
plt.axvline(x=0,color='k')
plt.grid()
plt.plot(x,y)
```
